#**QUESTION 1**
**What is the average compensation of associate professors at Auburn University-Main in Alabama?**

```python
import pandas as pd
df = pd.read_excel('aaup.xlsx', index_col=0)
auburn_data = df[(df['College name'] == 'Auburn University- Main') & (df['State (postal code)'] == 'AL')]
avg_compensation = auburn_data.loc[:, 'Average compensation - associate professors'].values[0]
print(f"The average compensation of associate professors at Auburn University-Main in Alabama is {avg_compensation}.")
```

#**Question 2**
**What is the distribution of average compensation for all ranks in each state?**
**Show it in the bar graph**



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel('aaup.xlsx')
avg_comp_by_state = df.groupby('State (postal code)')['Average compensation - all ranks'].mean()
plt.hist(avg_comp_by_state.values, bins=10)
plt.title('Distribution of Average Compensation by State')
plt.xlabel('Average Compensation')
plt.ylabel('Frequency')
plt.show()
```

#**Question 5**
**Create a PMF and CDF graph for the distribution of average salaries of all professors in Alabama universities.**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('aaup.xlsx')

alabama_df = df[df['State (postal code)'] == 'AL']

salaries = alabama_df['Average salary - all ranks']

pmf = salaries.value_counts(normalize=True).sort_index()
plt.bar(pmf.index, pmf.values)
plt.title('PMF for Average Salaries of All Professors in Alabama Universities')
plt.xlabel('Salary')
plt.ylabel('PMF')
plt.show()

cdf = pmf.cumsum()
plt.step(cdf.index, cdf.values)
plt.title('CDF for Average Salaries of All Professors in Alabama Universities')
plt.xlabel('Salary')
plt.ylabel('CDF')
plt.show()
```

#**QUESTION 8**
**What is the breakdown of the top 10% of high school class for new students enrolled in colleges? Show it in the pie chart form**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('usnews.xlsx')
df['Pct. new students from top 10% of H.S. class'] = pd.to_numeric(df['Pct. new students from top 10% of H.S. class'], errors='coerce')
top_10 = df['Pct. new students from top 10% of H.S. class']
above_10 = len(top_10[top_10 > 0])
below_10 = len(top_10) - above_10
counts = [above_10, below_10]
print(counts)
labels = ['Above 10%', 'Below 10%']
plt.pie(counts, labels=labels, autopct='%1.1f%%')
plt.title('Top 10% of H.S. Class')
plt.show()
```

#**QUESTION 9**
**Calculate and plot the probability mass function (PMF) of new students enrolled in albama college**


```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("usnews.xlsx")
alabama_data = data[data["State (postal code)"] == "AL"]

enrolled = alabama_data["Number of new students enrolled"]
enrolled_pmf = enrolled.value_counts(normalize=True)
enrolled_pmf.plot(kind="bar")
plt.xlabel("Number of new students enrolled")
plt.ylabel("PMF")
plt.title("PMF of Number of New Students Enrolled in Alabama Colleges")
plt.show()
```

#**QUESTION 10**
**Calculate and plot the cumulative distribution function (CDF) of in-state tuition costs for all colleges.**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel("usnews.xlsx")
instate_tuition = data["In-state tuition"].replace('*', np.nan).dropna().astype(float)
instate_tuition_cdf = np.cumsum(np.histogram(instate_tuition, bins=100, density=True)[0])

plt.plot(instate_tuition_cdf)
plt.title("CDF of In-state Tuition Costs")
plt.xlabel("Tuition Cost")
plt.ylabel("CDF")
plt.show()
```
